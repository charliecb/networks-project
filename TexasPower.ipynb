# Import Libraries

```python
# import libraries
import gurobipy as gp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import Data

```python
# import data
links = pd.read_csv("links.csv")
nodes = pd.read_csv("nodes.csv")
loads = pd.read_csv("loads.csv")
```

# Plot Nodes

```python
# plot nodes
plt.scatter(nodes["Longitude"], nodes["Latitude"], marker = ".")
plt.show()
```

Plot Links

```python
joined = links.merge(nodes, left_on = "Node1", right_on = "Node")
joined = joined.merge(nodes, left_on = "Node2", right_on = "Node")

start_lat = joined["Latitude_x"]
start_long = joined["Longitude_x"]
end_lat = joined["Latitude_y"]
end_long = joined["Longitude_y"]

for i in range(len(start_lat)):
    x = [start_long[i], end_long[i]]
    y = [start_lat[i], end_lat[i]]
    
    plt.plot(x, y)
    
plt.scatter(nodes["Longitude"], nodes["Latitude"], marker = ".")

plt.show()
```
